```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import os
epochs = 30
classes = ['pen', 'eraser', 'sharpener']
num_classes = len(classes)
experiments = [
    {"name": "baseline",    "lr": 0.001,   "batch_size": 16, "dropout": 0.5, "optimizer": "adam"},
    {"name": "high lr",     "lr": 0.1,     "batch_size": 16, "dropout": 0.5, "optimizer": "adam"},
    {"name": "low lr",      "lr": 0.00001, "batch_size": 16, "dropout": 0.5, "optimizer": "adam"},
    {"name": "no dropout",  "lr": 0.001,   "batch_size": 16, "dropout": 0.0, "optimizer": "adam"},
    {"name": "sgd",         "lr": 0.001,   "batch_size": 16, "dropout": 0.5, "optimizer": "sgd"},
    {"name": "small batch", "lr": 0.001,   "batch_size": 2,  "dropout": 0.5, "optimizer": "adam"},
]
transform = transforms.Compose([
    transforms.Resize((64, 64)),
    transforms.ToTensor(),
    transforms.Normalize([0.5, 0.5, 0.5], [0.5, 0.5, 0.5])
])
class SmallCNN(nn.Module):
    def __init__(self, dropout):
        super(SmallCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 16, 3, padding=1)
        self.relu = nn.ReLU()
        self.bn1 = nn.BatchNorm2d(16)
        self.pool = nn.MaxPool2d(2)
        self.conv2 = nn.Conv2d(16, 32, 3, padding=1)
        self.bn2 = nn.BatchNorm2d(32)
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(32 * 16 * 16, 128)
        self.dropout = nn.Dropout(dropout)
        self.fc2 = nn.Linear(128, num_classes)

    def forward(self, x):
        x = self.pool(self.bn1(self.relu(self.conv1(x))))
        x = self.pool(self.bn2(self.relu(self.conv2(x))))
        x = self.flatten(x)
        x = self.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        return x
all_train_losses = {}
all_val_losses = {}
all_val_accs = {}
dataset = datasets.ImageFolder("dataset/", transform=transform)
train_size = int(len(dataset) * 0.8)
val_size = len(dataset) - train_size
for exp in experiments:
    print(f"\n--- Running: {exp['name']} ---")

    train_ds, val_ds = torch.utils.data.random_split(dataset, [train_size, val_size])
    train_loader = DataLoader(train_ds, batch_size=exp["batch_size"], shuffle=True)
    val_loader = DataLoader(val_ds, batch_size=exp["batch_size"])

    model = SmallCNN(dropout=exp["dropout"])
    criterion = nn.CrossEntropyLoss()

    if exp["optimizer"] == "adam":
        optimizer = optim.Adam(model.parameters(), lr=exp["lr"])
    else:
        optimizer = optim.SGD(model.parameters(), lr=exp["lr"], momentum=0.9)

    train_losses = []
    val_losses = []
    val_accs = []

    for epoch in range(epochs):
        model.train()
        running_loss = 0
        for images, labels in train_loader:
            optimizer.zero_grad()
            predictions = model(images)
            loss = criterion(predictions, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()

        train_losses.append(running_loss / len(train_loader))

        model.eval()
        val_loss = 0
        correct = 0
        total = 0
        with torch.no_grad():
            for images, labels in val_loader:
                predictions = model(images)
                val_loss += criterion(predictions, labels).item()
                _, predicted_classes = torch.max(predictions, 1)
                correct += (predicted_classes == labels).sum().item()
                total += labels.size(0)

        val_losses.append(val_loss / len(val_loader))
        accuracy = correct / total
        val_accs.append(accuracy)
        print(f"  Epoch {epoch+1}/{epochs}  train_loss: {train_losses[-1]:.4f}  val_loss: {val_losses[-1]:.4f}  acc: {accuracy:.2%}")

    all_train_losses[exp["name"]] = train_losses
    all_val_losses[exp["name"]] = val_losses
    all_val_accs[exp["name"]] = val_accs
os.makedirs("results", exist_ok=True)
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for name in all_train_losses:
    axes[0].plot(all_train_losses[name], label=name)
axes[0].set_title("Train Loss")
axes[0].set_xlabel("Epoch")
axes[0].set_ylabel("Loss")
axes[0].legend()
for name in all_val_losses:
    axes[1].plot(all_val_losses[name], label=name)
axes[1].set_title("Val Loss")
axes[1].set_xlabel("Epoch")
axes[1].set_ylabel("Loss")
axes[1].legend()
for name in all_val_accs:
    axes[2].plot(all_val_accs[name], label=name)
axes[2].set_title("Accuracy")
axes[2].set_xlabel("Epoch")
axes[2].set_ylabel("Accuracy")
axes[2].legend()
plt.tight_layout()
plt.savefig("results/comparison_all.png")
plt.show()
print("Saved to results/comparison_all.png")
```
